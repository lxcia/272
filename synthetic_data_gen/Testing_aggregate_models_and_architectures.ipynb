```python
import warnings
warnings.filterwarnings("ignore")
logregs = []
mlps = []
all_data = []
for i in range(15):
  df = pd.read_csv(f'clinic_datasets_sophie_hormone_generation/clinic_{i}.csv')
  all_data.append(df)
  X = df.drop(columns=['response_type', 'treatment'])
  y = df['treatment']
  clf = LogisticRegression(random_state=0).fit(X, y);
  print(f'Clinic {i} logreg accuracy: {clf.score(X, y)}')
  logregs.append(clf.score(X, y))
  clf = MLPClassifier(hidden_layer_sizes=(24)).fit(X, y)
  print([coef.shape for coef in clf.coefs_])
  print([inter.shape for inter in clf.intercepts_])
  print(f'Clinic {i} MLP accuracy: {clf.score(X, y)}')
  print(f'MLP LOSS: {log_loss(y, clf.predict_proba(X))}')
  mlps.append(clf.score(X, y))

print(f"Avg logreg accuracy: {np.average(logregs)}")
print(f"Avg MLP accuracy: {np.average(mlps)}")
```
